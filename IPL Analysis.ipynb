```python
def bowler_vs_batsman1(bowler,batsman):
    new_delivery=delivery[delivery['bowler']==bowler]
    return(new_delivery.groupby('batsman')['batsman_runs'].agg('sum').sort_values(ascending=False)[batsman])
```

```python
bowler_vs_batsman1('R Ashwin','GJ Maxwell')
```

```python
def bowler_vs_batsman2(bowler,batsman):
    new_delivery=delivery[delivery['bowler']==bowler]
    runs=new_delivery.groupby('batsman')['batsman_runs'].agg('sum').sort_values(ascending=False)[batsman]
    balls=new_delivery.groupby('batsman')['batsman_runs'].agg('count').sort_values(ascending=False)[batsman]
    sr=(runs/balls)*100
    print("Balls bowled :",balls)
    print("Runs scored :",runs)
    print("Strike Rate :",sr)
```

```python
bowler_vs_batsman2('R Ashwin','GJ Maxwell')
```

```python
bowler_vs_batsman2('JJ Bumrah','GJ Maxwell')
```
